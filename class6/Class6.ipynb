```python
#Create dataframe from a dictionary

import pandas as pd
import numpy as np

employees = {
    'Emp_ID': [101,102,103,104,105],
    'Name': ["hamza",'Talha','jawad','Honda','yamaha'],
    'Age': [24,56,30, np.nan, 29],
    'Salary': [50000, 60000,750000, 80000, np.nan],
    'Department': ['HR','IT','Finance', 'IT','HR'],
    'Location' :['NewYork','islamabad', 'Karachi', 'Lahore', 'Hyderabad']
}
# print(employees)
df = pd.DataFrame(employees)
print(df)
```

```python
#Handing Missing data
df['Age'].fillna(df['Age'].mean(), inplace=True)
# df['Salary'].fillna(df['Salary'].mean(), inplace=True)
df.dropna(subset=['Salary'], inplace=True)

print(df)
```

```python
#filter Employees
it_employees = df[df['Department'] == 'IT']
print('IT Emplyees:]\n ',it_employees)
```

```python
#Task
#Filter the employees who are working in Finance department
```

```python
#Sorting employees by Salary

df['Expereince'] = np.where(df['Salary'] > 70000, 'Senior', 'Junior')
# print(df)
sorted_df = df.sort_values(by='Salary', ascending=False)
print(sorted_df)
```

```python
#Creating another dataframe for merging
new_hires = {
    'Emp_ID': [106,107],
    'Name': ["Frank",'John'],
    'Age': [44,30],
    'Salary': [55000, 65000],
    'Department': ['HR','Finance'],
    'Location' :['France','Germany']

}
df_newhire = pd.DataFrame(new_hires)
print(df_newhire)
```

```python
df_final = pd.concat([sorted_df, df_newhire], ignore_index=True)
print(df_final)
```

```python
df_final['Experience'] = np.where(df_final['Salary'] > 70000, 'Senior', 'Junior')
print(df_final)
```

```python
#matplotlib
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.plot(x,y, marker='o', linestyle='--', color='r', label = 'Line1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Line Plot')
plt.legend()
plt.show()
```

```python
x = np.linspace(0,10,10)
# print(x)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10,6))

plt.plot(x,y1,color='red',linestyle='-',marker='o',markerfacecolor = 'yellow', label='Sin(x)')
plt.plot(x,y2,color='blue',linestyle='--',marker='x',label='Cos(x)')

plt.xlabel('X-axis',fontsize=12)
plt.ylabel('Y-axis', fontsize=12)
plt.title('Sin(x) and Cos(x)', fontsize=14)
plt.legend(loc='upper left')
plt.grid(True)
plt.annotate('Peak',xy=(2.57,1), xytext=(3,1.2), arrowprops={'facecolor':'black','shrink':0.05})

plt.show()
```

```python
#Sacatter Plot
x = np.random.rand(50)
y = np.random.rand(50)

plt.scatter(x,y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
categories = np.random.randint(0,3,50)
colors = ['red','green','blue']

plt.scatter(x,y, c=[colors[i] for i in categories], marker = 'o')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
```

```python
markers = ['o','s','D']

for i, marker in enumerate(markers):
    plt.scatter(x[categories == i], y[categories == i], marker=marker, label=f'Category {i}',alpha=0.1)

plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show
```

```python
z = np.random.rand(50) * 100  # Another variable for scaling
plt.scatter(x, y, s=z, alpha=0.5, c=z, cmap='coolwarm')
plt.colorbar(label="Size Scaling Variable")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Dynamic Scaling and Colormap")
plt.show()

```

```python
# Generate a linear relationship
x = np.linspace(0, 10, 50)
y = 3*x + np.random.randn(50) * 5  # y = 3x + noise
plt.scatter(x, y, color='blue', label="Data")
# Fit a linear trend line
m, b = np.polyfit(x, y, 1)  # First-degree polynomial fit
plt.plot(x, m*x + b, color='red', label="Trend Line")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.title("Scatter Plot with Regression Trend Line")
plt.show()

```

```python
import seaborn as sns
sns.regplot(x=x, y=y, scatter_kws={'color':'blue'}, line_kws={'color':'red'})
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot with Seaborn Regression")
plt.show()

```
